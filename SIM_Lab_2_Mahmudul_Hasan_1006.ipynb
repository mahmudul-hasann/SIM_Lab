```python
#2.Perform addition, subtraction, and multiplication operations on these matrices.
import numpy as np


matrix1 = np.random.randint(1, 10, size=(3, 3))
matrix2 = np.random.randint(1, 10, size=(3, 3))

print("Matrix 1:")
print(matrix1)

print("\nMatrix 2:")
print(matrix2)

# Matrix Addition
addition = matrix1 + matrix2
print("\nMatrix Addition:")
print(addition)

# Matrix Subtraction
subtraction = matrix1 - matrix2
print("\nMatrix Subtraction:")
print(subtraction)

# Matrix Multiplication (Dot Product)
multiplication = np.dot(matrix1, matrix2)
print("\nMatrix Multiplication:")
print(multiplication)
```

```python
#1.Compute the determinant, inverse, and rank of one of the matrices.
import numpy as np


matrix1 = np.array([[2, 4, 1],
                    [3, 1, 5],
                    [1, 2, 3]])

print("Matrix 1:")
print(matrix1)

# Determinant
det = np.linalg.det(matrix1)
print("\nDeterminant of Matrix 1:", det)


if det != 0:
    inverse = np.linalg.inv(matrix1)
    print("\nInverse of Matrix 1:")
    print(inverse)
else:
    print("\nMatrix 1 has no inverse (determinant is 0)")

# Rank
rank = np.linalg.matrix_rank(matrix1)
print("\nRank of Matrix 1:", rank)
```

```python
#2.Generate and Plot Two Random Vectors:
import numpy as np
import matplotlib.pyplot as plt
A_flat = np.random.rand(15)
B_flat = np.random.rand(15)
x = np.arange(1, 16)
plt.figure(figsize=(10, 5))
plt.plot(x, A_flat, marker='o', linestyle='-',label=' Matrix A')
plt.plot(x, B_flat, marker='s', linestyle='--',label=' Matrix B')
plt.title("Comparison of Two Random Vectors")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#2.Generate and Visualize a Random Matrix:
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
matrix = np.random.rand(4, 4)
plt.figure(figsize=(6, 5))
sns.heatmap(matrix, annot=True, cmap="YlGnBu", fmt=".2f",
            xticklabels=[f'Col {i+1}' for i in range(4)],
            yticklabels=[f'Row {i+1}' for i in range(4)])
plt.title("Heatmap of 4x4 Random Matrix")
plt.tight_layout()

plt.show()
```

```python
#3.Matrix Operations and Visualization:
import numpy as np
import matplotlib.pyplot as plt
matrix_a = np.random.rand(4, 4)
matrix_b = np.random.rand(4, 4)
add_result = matrix_a + matrix_b
sub_result = matrix_a - matrix_b
mul_result = np.dot(matrix_a, matrix_b)

def plot_matrix_bars(matrix, title):
    flat = matrix.flatten()
    plt.figure(figsize=(8, 4))
    plt.bar(range(1, len(flat) + 1), flat, color='skyblue')
    plt.title(title)
    plt.xlabel("Element Index")
    plt.ylabel("Value")
    plt.tight_layout()
    plt.show()

plot_matrix_bars(add_result, "Matrix Addition Result")
plot_matrix_bars(sub_result, "Matrix Subtraction Result")
plot_matrix_bars(mul_result, "Matrix Multiplication Result")
```
